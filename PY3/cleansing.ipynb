```python
dhoteles=pd.read_csv('C:\\repos\\hoteles\\DATA\\TRATADOS\\data_comentariosv1.csv',sep=';')
dhoteles
```

```python
##1. Limpiar minusculas
dhoteles.com_positivo = dhoteles.com_positivo.apply(lambda x: x.lower())
dhoteles['com_negativo']=dhoteles['com_negativo'].str.lower()
dhoteles['com_res']=dhoteles['com_res'].str.lower()

##2. Limpiar puntuacion
dhoteles['com_positivo'] = dhoteles['com_positivo'].str.replace('[^\w\s]','')
dhoteles['com_negativo'] = dhoteles['com_negativo'].str.replace('[^\w\s]','')
dhoteles['com_res'] = dhoteles['com_res'].str.replace('[^\w\s]','')

##3. Limpiar acentos
dhoteles['com_positivo']=dhoteles['com_positivo'].str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
dhoteles['com_negativo']=dhoteles['com_negativo'].str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
dhoteles['com_res']=dhoteles['com_res'].str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
```
